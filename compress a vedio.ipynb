```python
# Importing the module
from moviepy.editor import *
from IPython.display import HTML
import os
```

```python
# uploading the video we want to edit
video = VideoFileClip("VID-20191126-WA0186.mp4")
```

```python
# Where the new file will be stored
compressed_file = "video_resized.mp4"
```

```python
# getting width and height of video 1
width_of_video1 = video.w
height_of_video1 = video.h
```

```python
# print the quality of the original vedio
print("Width and Height of original video : ", end = " ")
print(str(width_of_video1) + " x ", str(height_of_video1))
```

```python
# get size of the original video
original_size = os.path.getsize("VID-20191126-WA0186.mp4")
print("Original video size:", original_size, "bytes")
```

```python
# compressing by 20%
video_resized = video.resize(0.2)
```

```python
# compress the video with H.264 codec and a bitrate of 5000 kbps
video_resized.write_videofile(compressed_file, codec='libx264', bitrate='5000')
```

```python
# getting width and height of video 2 which is resized
width_of_video2 = video_resized.w
height_of_video2 = video_resized.h
```

```python
# print the dimensions after compresssion
print("Width and Height of resized video : ", end = " ")
print(str(width_of_video2) + " x ", str(width_of_video2))
```

```python
# maxduration parameter take the max number of the secs to be compressed (no. of the min * 60)
video_resized.ipython_display(width= 480 , maxduration =320)
```

```python
# get size of the compressed video
compressed_size = os.path.getsize("video_resized.mp4")
print("Compressed video size:", compressed_size, " bytes")
```
